```python
%matplotlib inline
import sys
import numpy as np
import matplotlib.pyplot as plt
from humanposer.io import load_obj
from humanposer.uv_util import map_vertex_to_texture
fname = "bodymodels/smplx_uv.obj"

vertices, uv_coords, vertex_faces, texture_faces = load_obj(fname)


vertex_to_texture = map_vertex_to_texture(vertex_faces, texture_faces)


vertices.shape, vertices.dtype
uv_coords.shape, uv_coords.dtype
vertex_faces.shape, vertex_faces.dtype

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)

ax.scatter(uv_coords[:, 0], uv_coords[:, 1], s=1)

plt.show()
```
